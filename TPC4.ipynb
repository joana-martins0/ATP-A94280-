## Resumo:
Inicialmente, chama-se um menu com 10 opções para o utilizador escolher, ao escolher a ultinha opçao (0) o menu deixa de ser corrido.

```python
import random

def menu():
    print(""""As opções disponiveis sao:
1) Criar Função
2) Simplificar a Fração
3) Somar Frações
4) Subtrair Frações
5) Multiplicar Frações
6) Dividir Frações
7) Gerar lista de n Frações
8) Somar as Frações de uma Lista
9) Maior Fração de uma Lista
0) Sair
""")

def criarFracao(numerador,denominador):
    return (numerador,denominador)

def verFracao(f):
    print(str(int(f[0]))+"/"+str(int((f[1]))))

def mdc(a,b):
    if a<b:
        return mdc(b,a)
    elif a%b==0:
        return b
    else:
        return mdc(a,a%b) 

def somaFrac(f1,f2):
    a=(f1[0]*f2[1]+f2[0]*f1[1],f1[1]*f2[1])
    return simplificarFracao(a)

def simplificarFrac(f):
    num,denom=f
    a=mdc(num,denom)
    return (num/a,denom/a)

def subtrairFrac(f1,f2):
    a=(f1[0]*f2[1]-f2[0]*f1[1],f1[1]*f2[1])
    return simplificarFracao(a)

def dividirFrac(f1,f2):
    f3=(f2[1],f2[0])
    a=multiplicarFracao(f1,f3)
    return simplificarFracao(a)

def multiplicarFrac(f1,f2):
    a=(f1[0]*f2[0],f1[1]*f2[1])
    return simplificarFracao(a)

def maiorFracao(l):
    maior=(l[0][0],l[0][1])
    maior2=l[0][0]/l[0][1]
    for i in range(len(l)):
        div=l[i][0]/l[i][1]
        if div>maior2:
            maior2=l[i][0]/l[i][1]
            maior=(l[i][0],l[i][1])
    return maior

def randomFrac(n):
    l=[]
    for i in range(n):
        numerador=random.randint(1,10)
        denominador=random.randint(1,10)
        f=(numerador,denominador)
        l.append(f)
    return l

opcao="1"
while opcao!="0":
    menu()
    opcao=input("Qual a sua opção?")
    if opcao == "1":
        numerador=int(input("Diga o numerador"))
        denominador=int(input("Diga o denominador"))
        criarFracao(numerador, denominador)
    elif opcao== "2":
        f=(numerador,denominador)
        verFracao(simplificarFracao(f))
    elif opcao=="0":
        print("Ate à proxima")
    elif opcao=="3":
        numerador=int(input("Diga o numerador:"))
        denominador=int(input("Diga o denominador:"))
        f1=criarFracao(numerador, denominador)
        numerador2=int(input("Diga o numerador:"))
        denominador2=int(input("Diga o denominador:"))
        f2=criarFracao(numerador2, denominador2)
        verFracao(somaFracao(f1,f2))
    elif opcao=="4":
        numerador=int(input("Diga o numerador:"))
        denominador=int(input("Diga o denominador:"))
        f1=criarFracao(numerador, denominador)
        numerador2=int(input("Diga o numerador:"))
        denominador2=int(input("Diga o denominador:"))
        f2=criarFracao(numerador2, denominador2)
        verFracao(subtrairFracao(f1,f2))
    elif opcao=="5":
        numerador=int(input("Diga o numerador:"))
        denominador=int(input("Diga o denominador:"))
        f1=criarFracao(numerador, denominador)
        numerador2=int(input("Diga o numerador:"))
        denominador2=int(input("Diga o denominador:"))
        f2=criarFracao(numerador2, denominador2)
        verFracao(multiplicarFracao(f1,f2))
    elif opcao=="6":
        numerador=int(input("Diga o numerador:"))
        denominador=int(input("Diga o denominador:"))
        f1=criarFracao(numerador, denominador)
        numerador2=int(input("Diga o numerador:"))
        denominador2=int(input("Diga o denominador:"))
        f2=criarFracao(numerador2, denominador2)
        verFracao(dividirFracao(f1,f2))
    elif opcao=="7":
        n=int(input("Numero de frações:"))
        l=randomFracao(n)
        for j in range(len(l)):
            verFracao(l[j])
    elif opcao=="8":
        res=(0,1)
        for fracao in range(len(l)):
            res=somaFracao(l[fracao],res)
        verFracao(res)
    elif opcao=="9":
        m=maiorFracao(l)
        verFracao(m)
    else:
        print("Opção invàlida")
```
